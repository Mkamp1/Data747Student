Create a bar plot using Seaborn to visualize the count of flowers in each species (target) in the Iris dataset:

```python
import seaborn as sns

# Load iris dataset
iris = sns.load_dataset("iris")

# Create bar plot
sns.countplot(data=iris, x="species")
```

 Use Plotly to create an interactive scatter plot to visualize the relationship between the sepal length and sepal width of the flowers in the Iris dataset:

```python
import plotly.express as px

# Load iris dataset
iris = px.data.iris()

# Create scatter plot
fig = px.scatter(iris, x="sepal_width", y="sepal_length", color="species", hover_data=["petal_width", "petal_length"])

# Add tooltips
fig.update_traces(hovertemplate="Sepal Width: %{x}<br>Sepal Length: %{y}<br>Petal Width: %{customdata[0]}<br>Petal Length: %{customdata[1]}")

# Add axis labels and title
fig.update_layout(
    xaxis_title="Sepal Width",
    yaxis_title="Sepal Length",
    title="Relationship between Sepal Width and Sepal Length"
)

# Show the plot
fig.show()
```

Create a histogram using Matplotlib to visualize the distribution of petal lengths in the Iris dataset:

```python
import matplotlib.pyplot as plt

# Load iris dataset
iris = sns.load_dataset("iris")

# Create histogram
plt.hist(iris["petal_length"], bins=10, color="skyblue", alpha=0.8)

# Add labels and title
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.title("Distribution of Petal Length")

# Display the plot
plt.show()
```

Use Seaborn to create a box plot to compare the distribution of petal widths across different species (target) in the Iris dataset:

```python
import seaborn as sns

# Load iris dataset
iris = sns.load_dataset("iris")

# Create box plot
sns.boxplot(data=iris, x="species", y="petal_width")

# Add labels and title
plt.xlabel("Species")
plt.ylabel("Petal Width")
plt.title("Distribution of Petal Width by Species")

# Display the plot
plt.show()
```

 Create a line plot using Plotly to display the trend of petal widths over the different species (target) in the Iris dataset:

```python
import plotly.express as px

# Load iris dataset
iris = px.data.iris()

# Create line plot
fig = px.line(iris, x="species_id", y="petal_width", color="species", markers=True, title="Petal Width by Species")

# Display the plot
fig.show()
```

Use Matplotlib to create a stacked area plot to visualize the contribution of different species (target) to the total number of flowers in the Iris dataset:

```python
import matplotlib.pyplot as plt

# Load iris dataset
iris = sns.load_dataset("iris")

# Count flowers by species
species_count = iris["species"].value_counts()

# Create stacked area plot
plt.stackplot(species_count.index, species_count.values, labels=species_count.index)

# Add labels and title
plt.xlabel("Species")
plt.ylabel("Count")
plt.title("Contribution of Species to Total Number of Flowers")

# Add legend
plt.legend(loc="upper left")

# Display the plot
plt.show()
```

Create a heat map using Seaborn to visualize the correlation matrix of numerical variables (sepal length, sepal width, petal length, petal width) in the Iris dataset:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load iris dataset
iris = sns.load_dataset("iris")

# Extract numerical columns for correlation matrix
numerical_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
iris_numerical = iris[numerical_columns]

# Calculate correlation matrix
correlation_matrix = iris_numerical.corr()

# Create heat map
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Add title
plt.title('Correlation Matrix')

# Show the plot
plt.show()
```
