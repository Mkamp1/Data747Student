# Load and parse SAS input program

```python
import re
import pandas as pd

# Path to your SAS Input Program
sas_input_path = "/mnt/data/2019XXH-SAS-Input-Program.sas"

# Read SAS input program
with open(sas_input_path, "r", errors="ignore") as f:
    sas_lines = f.readlines()

colspecs = []
names = []

pattern = re.compile(r"@(\d+)\s+(\w+)\s+(\d+)\.")

for line in sas_lines:
    match = pattern.search(line)
    if match:
        start = int(match.group(1)) - 1        # convert SAS 1-based â†’ python 0-based
        width = int(match.group(3))
        end = start + width
        colspecs.append((start, end))
        names.append(match.group(2))

len(colspecs), len(names)
```

# Load the fixed-width YRBSS .dat file

```python
data_path = "/mnt/data/XXH2019_YRBS_Data.dat"

df = pd.read_fwf(
    data_path,
    colspecs=colspecs,
    names=names
)

df.head()
```
