## Data Understanding

The dataset used in this project is the 2019 National Youth Risk Behavior Surveillance System (YRBSS), collected by the Centers for Disease Control and Prevention (CDC). The YRBSS is a nationwide survey designed to monitor a wide range of health-related behaviors among U.S. high school students. These behaviors include mental health indicators, substance use, injury-related behaviors, sexual health, physical activity, dietary habits, and other factors affecting adolescent well-being.

The 2019 survey includes over 13,000 respondents and represents a statistically weighted national sample of students in grades 9–12 across the United States.

The raw YRBSS dataset is provided in fixed-width (.dat) format and must be read using column definitions supplied in a CDC SAS input script. After applying these definitions, the resulting dataset includes approximately 134 variables, each corresponding to a specific survey question or derived value.

Example Variable Categories
Q6–Q15 — Demographics (age, gender, grade, and race/ethnicity).
QNOBESE / BMI variables — Height, weight, BMI, obesity indicators.
QNDEP / QNANX — Mental-health measures (feeling sad, anxiety indicators).
QNWATER, QNSODA — Dietary behaviors.
QNSMOKE / QNALCSIP — Tobacco and alcohol use.
QNFIGHT / QNBULLY — Violence- and injury-related behaviors.
PSU, STRATUM, WEIGHT — Survey sampling and weighting variables.
Because YRBSS is a probability-sample survey, records must be interpreted using sampling weights to ensure national representativeness. The dataset contains no personally identifiable information, making it appropriate for academic use.

Missing Values
Although the fixed-width file does not contain blank entries, many variables include coded missing values, such as:

7 — Refused
8 — Not sure
9 — Missing / Not applicable
These values must be cleaned before modeling.

```python
# Load Data

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import re

data_file = "yrbs2019.dat"
sas_layout = "yrbs2019_input.sas"

# Custom parser for SAS INPUT pointer format (@start var width.)

def parse_sas_input_pointer_format(file_path):
    colspecs = []
    names = []
    
# Pattern example: @17 Q1 $1.   or   @29 Q6 4.2
    pattern = re.compile(r"@(\d+)\s+(\w+)\s+(\$?)(\d+)")
    
    with open(file_path, "r") as f:
        lines = f.readlines()
    
    entries = []
    
    for line in lines:
        match = pattern.search(line)
        if match:
            start = int(match.group(1)) - 1       # convert to 0-based
            name = match.group(2)
            width = int(match.group(4))           # width from SAS format
            end = start + width
            entries.append((start, end, name))
    
# sort by start position
    entries.sort(key=lambda x: x[0])
    
# build final lists
    for start, end, name in entries:
        colspecs.append((start, end))
        names.append(name)
    
    return colspecs, names


# Parse SAS File

colspecs, names = parse_sas_input_pointer_format(sas_layout)
print("Columns detected:", len(names))

data = pd.read_fwf(data_file, colspecs=colspecs, names=names)

data.head()

```

## 3. Data Preparation

This section describes the steps taken to prepare the 2019 YRBSS dataset for analysis.  
Because the CDC distributes the national YRBSS data in a fixed-width text format, several preprocessing steps were required to convert the raw ASCII file into a structured DataFrame. These steps include parsing the SAS input program, handling CDC-coded missing values, selecting relevant variables, converting data types, and performing initial cleaning prior to modeling.

The primary goals of the data preparation stage are to:

- **Ensure the dataset loads correctly** into a fully structured tabular format using the CDC’s SAS layout file.
- **Replace CDC-coded missing values** (7, 8, 9, 77, 88, 99, depending on the question) with standard `NaN` to support analysis.
- **Select a subset of variables** aligned with the project’s research question on youth mental and physical health indicators.
- **Convert variables to appropriate data types**, including categorical encoding for survey responses and numeric conversion for continuous measures.
- **Create a cleaned modeling DataFrame** that is suitable for training predictive models in Section 4.

The following subsections describe each data preparation step in detail, including loading the dataset using the SAS pointer-format parser, cleaning missing or invalid values, selecting analytical variables, and preparing the final DataFrame used for modeling and evaluation.


```python
# Data Preparation

import pandas as pd
import numpy as np

# Replace CDC missing codes with NaN
coded_missing = [7, 8, 9, 77, 88, 99]
data = data.replace(coded_missing, np.nan)

# Select variables for modeling
# Demographics
# Q2 = Sex, Q3 = Grade, Q6 = Age category, RACEETH = recoded race/ethnicity
# Mental health outcome (TARGET): QN8 = felt sad/hopeless ≥2 weeks
# Behaviors: alcohol, vaping, smoking, soda, physical activity, obesity, BMI
selected_columns = [
    'Q2',          # Sex
    'Q3',          # Grade
    'Q6',          # Age category
    'RACEETH',     # Race / Ethnicity

    'QN8',         # TARGET: depression indicator

    'QN12',        # Alcohol use
    'QNDAYEVP',    # Days used e-cigarettes
    'QNDAYCIG',    # Days smoked cigarettes
    'QNSODA1',     # Soda consumption
    'QNPA0DAY',    # Physically inactive (0 days)
    'QNOBESE',     # Obesity status
    'BMIPCT'       # BMI percentile
]

# Keep only columns that actually exist in the dataset (safety check)
selected_columns = [c for c in selected_columns if c in data.columns]
cleaned = data[selected_columns].copy()

print("Selected columns:", selected_columns)

# Convert demographics to categorical
categorical_vars = ['Q2', 'Q3', 'Q6', 'RACEETH']
for col in categorical_vars:
    if col in cleaned.columns:
        cleaned[col] = cleaned[col].astype('category')

# Ensure numeric variables are numeric
for col in cleaned.columns:
    if col not in categorical_vars:
        cleaned[col] = pd.to_numeric(cleaned[col], errors='coerce')

# Basic checks
print("\nCleaned data types:")
display(cleaned.dtypes)

print("\nMissing values per variable:")
display(cleaned.isnull().sum())

print("\nPreview of cleaned dataset:")
display(cleaned.head())

print("\nFinal cleaned shape (rows, columns):")
cleaned.shape

```

```python
# Drop Rows With Missing Values

cleaned_before = cleaned.shape[0]

cleaned = cleaned.dropna()

cleaned_after = cleaned.shape[0]

print(f"Rows before dropping missing: {cleaned_before}")
print(f"Rows after dropping missing:  {cleaned_after}")
print(f"Rows removed:                 {cleaned_before - cleaned_after}")

print("\nPreview of dataset after dropping missing values:")
display(cleaned.head())

cleaned.shape
```

## 4. Feature Engineering and Data Splitting

This section describes the steps used to transform the cleaned YRBSS dataset into a format suitable for supervised machine learning models. After completing data preparation and removing records with missing values, several additional transformations were required to prepare the data for modeling.

The YRBSS dataset contains several demographic variables that are categorical in nature, including sex (Q2), grade level (Q3), age category (Q6), and race/ethnicity (RACEETH). Since most machine learning algorithms require numerical input, all categorical variables were converted into numerical form using **one-hot encoding**. This transformation creates a binary indicator column for each category while preventing the introduction of unintended ordinal relationships.

One-hot encoding resulted in an expanded feature matrix, increasing the total number of predictor variables from the original 12 to 46 encoded features.

The target outcome for this project is **poor mental health**, operationalized using the YRBSS item **QN8**, which asks whether the student felt sad or hopeless almost every day for at least two weeks during the past year.  
- 1 = Yes (poor mental health)  
- 2 = No  
- Values coded 7, 8, or 9 were treated as missing and removed during data cleaning.

For modeling purposes, the target variable was recoded into a binary format:
- 1 = Poor mental health  
- 0 = No reported symptoms

After encoding and cleaning the dataset, the distribution of the target variable remained highly imbalanced:
- Only **≈5%** of students reported persistent sadness or hopelessness (QN8 = 1)
- **≈95%** reported no symptoms (QN8 = 0)

This imbalance is expected in national youth health surveys but presents challenges for model performance.  
Although techniques such as **SMOTE, ADASYN, or class-weighted models** could be applied, these methods were not covered in this course and were not used.  
Instead, model performance will be evaluated using metrics appropriate for imbalanced data, including:
- Precision  
- Recall  
- F1-score  
- ROC-AUC  

To evaluate model performance, the dataset was randomly divided into:
- **Training set:** 75% of observations  
- **Testing set:** 25% of observations  

Stratified sampling was used to ensure that the class distribution of the target variable (QN8) was preserved in both sets. This approach prevents the minority class from being underrepresented during training and ensures a more reliable assessment of model performance.

With the encoded, cleaned, and split dataset prepared, the next section will introduce and train several supervised learning models to identify key demographic and behavioral predictors of poor mental health among U.S. high school students.

## 5. Modeling

The goal of this section is to build supervised machine learning models that predict whether a student experienced poor mental health, operationalized as reporting persistent sadness or hopelessness for two or more weeks (QN8). Two classification algorithms covered in the course were selected:

- **Logistic Regression** – a linear baseline classifier that models the probability of the positive class.
- **Decision Tree Classifier** – a flexible, interpretable non-linear model capable of capturing complex interactions.

Both models were trained using the cleaned dataset prepared in Section 3 and the encoded feature set produced in Section 4. Model performance was evaluated using accuracy, the confusion matrix, and the classification report on the held-out testing set. Because SMOTE and advanced resampling were not covered in the course, the models were trained on the imbalanced dataset without oversampling, consistent with course expectations.

The following subsections present the training code, predictions, and evaluation metrics for each model.


```python
# Logistic Regression Model

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Create model
log_model = LogisticRegression(max_iter=200, solver="lbfgs")

# Train model
log_model.fit(X_train, y_train)

# Predict on test set
log_pred = log_model.predict(X_test)

# Evaluate performance
log_accuracy = accuracy_score(y_test, log_pred)
log_cm = confusion_matrix(y_test, log_pred)

print("Logistic Regression Accuracy:", round(log_accuracy, 4))
print("\nConfusion Matrix:\n", log_cm)
print("\nClassification Report:\n", classification_report(y_test, log_pred))


# Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier

# Create model
tree_model = DecisionTreeClassifier(
    criterion="gini", 
    max_depth=None, 
    random_state=42
)

# Train model
tree_model.fit(X_train, y_train)

# Predict
tree_pred = tree_model.predict(X_test)

# Evaluate
tree_accuracy = accuracy_score(y_test, tree_pred)
tree_cm = confusion_matrix(y_test, tree_pred)

print("Decision Tree Accuracy:", round(tree_accuracy, 4))
print("\nConfusion Matrix:\n", tree_cm)
print("\nClassification Report:\n", classification_report(y_test, tree_pred))

print("Model Comparison:")
print("------------------")
print("Logistic Regression:", round(log_accuracy, 4))
print("Decision Tree:", round(tree_accuracy, 4))
```
